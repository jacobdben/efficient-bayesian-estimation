```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from copy import deepcopy
```

```python
Tc = 10000
#fname = 'partial_update_map_T_' + str(Tc) + '.csv'
#df = pd.read_csv('partial_maps/'+fname).drop(['Unnamed: 0'], axis=1)
fname = 'map_no_dephasing.csv'
df = pd.read_csv('partial_maps/'+fname).drop(['Unnamed: 0'], axis=1)
N = 200
M = 500

fmap = np.empty((N,M,7))
fmap[:] = np.nan
```

```python
mu = df.to_numpy()[0,0]
j = 0
k = 0

count = 0

for i in range(df.shape[0]):
    
    if df.to_numpy()[i,0] == mu:
        k += 1
        
    if df.to_numpy()[i,0] != mu or i == df.shape[0]-1:
        mu = df.to_numpy()[i,0]
        
        # To k for whole dataset
        for l in range(0,k-1):
            fmap[j,-l-1,:] = deepcopy(df.to_numpy()[i-l-1,:])
            count += 1
        
        j += 1
        k = 1
print(count)
```

```python
# Impute incorrect values
fmap[0,:,3] = 0

fmap[0,:,2] = 1/np.sqrt(2/Tc**2 + fmap[0,:,1]**2)*2*0.87287287211

fmap[0,261:272,2:] = np.nan




fmap[0,52,2:] = np.nan
fmap[0,96,2:] = np.nan

#fmap[2,117,2:] = np.nan

```

```python
old_mu = []
old_sig = []
fixed_t = []
fixed_mu0 = []
fixed_sig0 = []
fixed_mu1 = []
fixed_sig1 = []

for i in range(fmap.shape[0]):
    for j in range(fmap.shape[1]):
        if not np.isnan(fmap[i,j,2]):
            
            old_mu.append(fmap[i,j,0])
            old_sig.append(fmap[i,j,1])
            fixed_t.append(fmap[i,j,2])
            fixed_mu0.append(fmap[i,j,3])
            fixed_sig0.append(fmap[i,j,4])
            fixed_mu1.append(fmap[i,j,5])
            fixed_sig1.append(fmap[i,j,6])
        

old_mu = np.array(old_mu)
old_sig = np.array(old_sig)       
fixed_t = np.array(fixed_t)
fixed_mu0 = np.array(fixed_mu0)
fixed_sig0 = np.array(fixed_sig0)
fixed_mu1 = np.array(fixed_mu1)
fixed_sig1 = np.array(fixed_sig1)

df_new = pd.DataFrame({'mu_old':old_mu, 'sig_old':old_sig, 't_opt':fixed_t, 
                       'mu0':fixed_mu0, 'sig0':fixed_sig0, 'mu1':fixed_mu1, 'sig1':fixed_sig1})
df_new
```

```python
df_new.to_csv('partial_maps/'+'fixed_'+fname)
```
